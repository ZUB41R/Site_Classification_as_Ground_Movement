# Predicting the Site Classification of Ground Movement (GM) using Deep Neural Network on Geospatial Datasets

The main objective of this project is to obtain the amount of `Ground Movement` of a construction site using `Machine Learning` algorithms. The ground truth values of the target `Ground Movement` was calculated from the `Atterberg Limits` of the soil samples collected from the site.

This algorithm is to predict the  ground movement in terms of Site Classification by using Geospatial datasets.

# 2. Import data
Import data with necessary pandas, numpy and plot libraries. After importing we separate the dataset into `feature` that are used to predict the `target` that we are after.

```python
# import necessary libraries

from __future__ import print_function
import numpy as np
import pandas as pd
from time import time
import matplotlib.pyplot as plt
from IPython.display import display # allows the use of display() for DataFrames

%matplotlib inline

#load the dataset cleaned from "Historical Perth Database" provided by UDREW
data = pd.read_csv("Spatial_radio_cleaned_reduced_features.csv")
print(data.shape)
#data= data[(data[['GM']]!=0).all(axis=1)]
print(data.shape)

#targets_raw = data[['GM']]
# Binning the data into site sub_classification
targets_raw = pd.DataFrame(data['GM'])
binInterval = [-1, 0, 10, 20, 30, 40, 50, 60, 75, np.inf]
binLabels = ['A', 'S_1', 'S_2', 'M_1', 'M_2', 'H1_1', 'H1_2', 'H2', 'E']
targets_raw_binned = pd.cut(data['GM'], bins = binInterval, labels=binLabels)
display(targets_raw_binned.head(5))


#isolate the dataset into features and target
features_raw = data.drop(['GM'], axis=1)
features_raw_Surface_Geology = features_raw.drop('Surface_Geology', axis=1)



# features that do not contain 'Lat', 'Lon', and 'Avg_D'
#features_raw = data[['USC', 'Consistency', 'Gravel_grading', 'Gravel_spherocity',
#                   'Sand_grading','Sand_spherocity']]


#features_raw['Lat'].describe()

#display the first 3 records of each dataset
print('\n','All Data:\n----------')
display(data.head(3))
print('\n','Feature Data:\n---------------')
display(features_raw.head(3))
print('\n','Target Data:\n---------------')
display(targets_raw.head(3))

display(features_raw_Surface_Geology.head(3))
data.describe()
display(data.columns)
display(len(data.columns))
data.describe()
```

# 3. Data processing
This step includes normalizing the numerical dataset and one-hot encoding for the classes of teh `features` defined as text.

***
### 3.1 Normalize each of the numerical `features` and `targets`
***

```python
# Import sklearn.preprocessing.StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler and then apply it to the features
scaler_1 = MinMaxScaler() # default=(0, 1)
scaler_2 = MinMaxScaler()

display(targets_raw_binned.head(5))
#numerical_targets = ['GM']
#numerical_features = ['Lat', 'Lon', 'Avg_D'] 

#display(log_targets.head(5))
# Transform targets
#targets_minmax_transform = pd.DataFrame(data=targets_raw)
#targets_minmax_transform[numerical_targets] = scaler_1.fit_transform(targets_raw[numerical_targets])
#targets_minmax_transform[numerical_targets] = scaler_1.fit_transform(taregets_raw[numerical_targets])

# Transforms features
features_minmax_transform = pd.DataFrame(data = features_raw)
features_minmax_transform[features_raw_Surface_Geology.columns] = scaler_2.fit_transform(features_raw_Surface_Geology)



# show an example of a record with scaling applied
display(features_minmax_transform.head(3))
#display(targets_minmax_transform.head(3))
```

***
### 3.2 One hot-encoding for features that have string variables
***

```python
# One-hot encode the 'features_minmax_transform'
features_final = pd.get_dummies(features_minmax_transform)
GM_encoded = pd.get_dummies(targets_raw_binned)

# Print the number of features after one-hot encoding
encoded = list(features_final.columns)
print("{} total features after one-hot encoding.".format(len(encoded)))
print(encoded)
display(features_final.head(10))
```

# 4. Build the Model

***
### 4.1 Shuffle and split data into training and testing subsets
***

```python
# import train test split
from sklearn.model_selection import train_test_split

# set the target as Ground Movement (GM)
#GM = targets_minmax_transform[numerical_targets]
GM = GM_encoded
display(GM.head(5))

# Split the 'features' and 'target' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_final, GM, test_size=0.2, random_state=0)

# Split the training set into training and validation subsets
#X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
                                                                   
# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
#print("Validation set has {} samples.".format(X_valid.shape[0]))
print(GM.count())

#y_valid.shape
display(y_train.head(5))
display(y_test.head(5))
#display(y_valid.head(5))
display(X_train.head(5))
```

***
### 4.2 Build the Neural Network architecture
***

```python
# Import the libraries
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation 
from keras.layers.normalization import BatchNormalization
from keras.layers import GlobalAveragePooling1D
# Building the model
model = Sequential()
model.add(BatchNormalization(input_shape=(45,), momentum=0.99, epsilon=0.001, center=True, 
                             scale=True, beta_initializer='zeros', gamma_initializer='ones', 
                             moving_mean_initializer='zeros', moving_variance_initializer='ones'))

model.add(Dense(45, activation='relu', input_shape=(45,), kernel_initializer='RandomNormal'))
model.add(Dropout(.1))
model.add(BatchNormalization())

model.add(Dense(90, activation='relu', kernel_initializer='RandomNormal'))
model.add(Dropout(.1))
model.add(BatchNormalization())

#model.add(Dense(180, activation='relu', kernel_initializer='RandomNormal'))
#model.add(Dropout(.1))
#model.add(BatchNormalization())

#model.add(Dense(360, activation='relu', kernel_initializer='RandomNormal'))
#model.add(Dropout(.1))
#model.add(BatchNormalization())


model.add(BatchNormalization())
model.add(Dense(9, activation='softmax'))


model.summary()

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='Adamax', metrics=['accuracy'])
```

***
### 4.3 Train the model
***

```python
from keras.callbacks import ModelCheckpoint

# Train the model
checkpointer = ModelCheckpoint(filepath='GM_NN_Site_Class.model.best.hdf5',
                              verbose=1, save_best_only=True)
hist = model.fit(X_train, y_train, batch_size=35, epochs=20,
                validation_split=0.2, callbacks=[checkpointer],
                verbose=1, shuffle=True)
```

***
### 4.4 Load the optimized weights
***

```python
model.load_weights('GM_NN_Site_Class.model.best.hdf5')
```


# 5. Model evaluation

***
### 5.1 Accuracy
***


```python
score = model.evaluate(X_test, y_test, verbose=1)
accuracy = score[1]

# Print test accuracy
print('Accuracy: %.4f' %accuracy)
```

***
### 5.2 Confusion matrix
***

```python
from sklearn import svm, datasets
from sklearn.metrics import confusion_matrix
%matplotlib inline
import matplotlib.pyplot as plt

# Plot a confusion matrix.
# cm is the confusion matrix, names are the names of the classes.
def plot_confusion_matrix(cm, names, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(names))
    plt.xticks(tick_marks, names, rotation=45)
    plt.yticks(tick_marks, names)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
```

```python
predict_GM = model.predict(X_test)
#print(predict_GM.shape)
predict_GM_ = np.argmax(predict_GM, axis=1)
#print(predict_GM_)

#print(y_test)
#print(y_test)
y_test_ = y_test.values
#print(y_test_)
y_test_2 = np.argmax(y_test_, axis=1)
#y_test_2 = df.columns.get_loc()

#y_test_2_ = df.columns.get_loc([y_test_2])
#print(y_test_2)



# Compute confusion matrix
cm = confusion_matrix(y_test_2, predict_GM_)
np.set_printoptions(precision=2)
print('Confusion matrix')
print(cm)
#plt.figure()
plot_confusion_matrix(cm, binLabels)


```

```python
# Import data that are used as input features for prediction
to_predict_df = pd.read_csv("Dry_data_Geospatial.csv")

# Assign the target
#GM_Lab = to_predict_df['GM']

# Drop the target from the dataset to separate the features
to_predict_df_ = to_predict_df.drop(['GM', 'Surface_Geology'], axis=1)

#Print both of the model inputs and new dataset inputs to check whether they are same
print("Imported features that are used to predict GM:\n.................")
display(to_predict_df_.head(3))
print("Input features of the model:\n.................")
display(features_raw.head(3))

# Scaling the numerical features of the to_predict dataset
to_predict_minmax_transform = pd.DataFrame(data=to_predict_df_)
to_predict_minmax_transform[to_predict_df_.columns] = scaler_2.fit_transform(to_predict_df_)

# Display the tansformed "to_predict" data
print("Scaled input features:\n------------")
display(to_predict_minmax_transform.head(3)) 

# Prepare the training and testing dataset to concatenate
train_df = features_minmax_transform
train_df['label'] = 'train'
to_predict_minmax_transform['label'] = 'predict'

# Concat
concat_df = pd.concat([train_df, to_predict_minmax_transform])

# Create dummies
features_df = pd.get_dummies(concat_df)

# Show encoded features
concat_encoded = list(features_df.columns)
print("{} total features after one-hot encoding.".format(len(concat_encoded)))
print(concat_encoded)

# Split the data
train_df = features_df[features_df['label_train'] == 1]
to_predict_df_ = features_df[features_df['label_predict'] == 1]

# drop the feature named as "label"
train_df = train_df.drop(['label_train', 'label_predict'], axis=1)
to_predict_df_ = to_predict_df_.drop(['label_train', 'label_predict'], axis=1)

# Display both of the datasets
print('Encoded features of the training set:')
display(train_df.head(3))
print('Encoded features from the to_predict set:')
display(to_predict_df_.head(3))
```

***
### 6.2 Prediction
***

```python
#Show prediction

# Predict GM using the model tuned in "5.1 Model Evaluation"
predict_GM = model.predict(to_predict_df_)

predict_GM_ = np.argmax(predict_GM, axis=1)
#print(predict_GM_)
#display(predict_GM.shape)

# List of the Site Classification
print("All sites\n.......\n\n{}".format(binLabels))

GM_Prediction_ = []
df = pd.DataFrame(binLabels)

for index, value in enumerate(predict_GM_):
    GM_Prediction_.append(df.values[value])

# Print the predicted GM values
print("\nPredicted Site Classification:\n.....................")
display(GM_Prediction_)
```
